```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,5,0.1)
y = x ** 3
y2 = y ** 5
# plt.plot(x,y, color = 'black', marker='*')
plt.plot(x,y, color = 'black', linewidth = 2)
plt.plot(x,y2, color = 'r')
plt.ylabel('y')
plt.xlabel('Range')
# plt.legend('r','y')
plt.title('Matplotlib demo')
plt.grid()
plt.text(1.89,89, 'Text')

```
